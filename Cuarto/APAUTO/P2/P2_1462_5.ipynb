## Justificacion del rendimiento de KNN

EL inconveniente de KNN viene cuando las variables no se encuentran normalizadas, esto sucede por ejemplo si tenemos un atributo que se muve en el (0, 1), y otro que se mueve en rangos (0, 1000). Cuando tenemos estos datos, la segunda variable va a influir mucho mas en el calculo de la distancia euclidea, y consecuentemente en la decion del modelo. A este suceso se le conoce como la maldicion de la dimensionalidad

Es por eso que se aplica una normalizacion de las varibles, para que los rangos por los que se mueven los valores sean parecidos.

En cuanto a nuestros ejemplos, se observa claramente que cuando los atributos no estan normalizados, el rendimiento cae, precisamente por lo que hemos comentado. Con que haya un atributo que se mueva en los intervalors (300-500 por ejemplo), ya esta influyendo mucho mas que el resto (si se mueve de 0 a 1 por ejemplo), y con lo cual los errores aumentan.


## Analizar el impacto  de K

Se puede observar que con K = 1, no es un caso muy optimo, ya que tan solo un valor, no va a determinar que la mayoria de los valores cercanos sean iguales a ese K predicho. Igual ese K predice 1, pero los otros 2 K siguientes mas cercanos son de calse 2. Esto se suele llamar sobreajuste de los datos

A medida que aumentamos la K (3, 5, ...) es cuando observamos un menor error, ya que la muestra es lo suficientemente pequeña para no sobreajustar los datos, pero a su vez no demasiado grande para suavizar los datos.

Sin embargo, cuando ascendemos a K mas grandes (11, 21, ...), iremos aumentando en errores, ya que aumentamos el area de puntos, y esto tiende a "suavizar" los resultados.

## Normalizando los datos

```python
# Funciones auxiliares para normalizar
def calcularMediasDesv(datos):
    mean_desv = []
    attr_cont = 0

    # Para cada columna (menos la clase)
    for attr in datos.columns[:-1]:
        # Se guarda la media y la desviacion tipica
        mean_desv.append((np.mean(datos[attr]), np.std(datos[attr])))
        attr_cont += 1

    return mean_desv

def normalizarDatos(datos):

    data = pd.DataFrame(datos)
    # Se llama a calcular las medias
    means_desv = calcularMediasDesv(data)

    # Se recorre al mismo tiempo los dos arrays (columnas, y media_desviacion),
    # y se aplica la formula d enormalizacion a todos los datos de la columna
    for attr, mean_desv in zip(data.columns[:-1], means_desv):
        data[attr] = [(value - mean_desv[0]) / mean_desv[1]
                        for value in data[attr]]

    return data.to_numpy()
```

```python
# Pruebas para datos normalizados

km_1_clusters_n = km_1_classifier.predict(normalizarDatos(dataset_iris.datos))
km_2_clusters_n = km_2_classifier.predict(normalizarDatos(dataset_iris.datos))
km_3_clusters_n = km_3_classifier.predict(normalizarDatos(dataset_iris.datos))
km_4_clusters_n = km_4_classifier.predict(normalizarDatos(dataset_iris.datos))
km_5_clusters_n = km_5_classifier.predict(normalizarDatos(dataset_iris.datos))

print("1n", km_1_clusters_n)
print("2n", km_2_clusters_n)
print("3n", km_3_clusters_n)
print("4n", km_4_clusters_n)
print("5n", km_5_clusters_n)
```

Si observamos, no hay diferencias notables por haber sido normalizados los datos, que no sean las tipicas por el proceso random de elección de los centroides.

Pero normalizar creemos que no es una buena solucion, porque si lo que buscamos son clusters de puntos esparcidos en el espacio y los normalizamos, lo que estamos haciendo es acercarlos más entre ellos, y eso puede empeorar el rendimiento del algoritmo

```python
def predictions(classifier: KNeighborsClassifier, X, y, n_iters):
    errores = []

    for _ in range(n_iters):
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
        classifier.fit(X_train, y_train)
        y_pred = classifier.predict(X_test)
        errores.append(1 - accuracy_score(y_test, y_pred))

    return np.mean(errores), np.std(errores)
```

```python
from sklearn.cluster import KMeans
from sklearn import metrics
from collections import Counter
```

```python
# Data
df_iris = pd.read_csv("./iris.csv")

X_iris = df_iris.iloc[:, :-1]
y_iris = df_iris.iloc[:, -1]

# Classifiers fits
km_1_classifier = KMeans(n_clusters=1, init='random', max_iter=10000, tol=0).fit(X_iris)
km_2_classifier = KMeans(n_clusters=2, init='random', max_iter=10000, tol=0).fit(X_iris)
km_3_classifier = KMeans(n_clusters=3, init='random', max_iter=10000, tol=0).fit(X_iris)
km_4_classifier = KMeans(n_clusters=4, init='random', max_iter=10000, tol=0).fit(X_iris)
km_5_classifier = KMeans(n_clusters=5, init='random', max_iter=10000, tol=0).fit(X_iris)
```

```python
# Esta funcion se enctarga de devolver K clusters, con la etiqueta correspondiente a cada uno
# de los elementos dentro de un cluster
def get_predicted_cluster_labels(predictions, k):
    clusters = [[] for _ in range(k)]

    # Se itera entre todos los elementos predichos, los cuales van en orden
    i = 0 # Indice
    for pred in predictions:
        # Mete en el cluster correspondiente la label actual de esa clase
        clusters[pred].append(y_iris[i])
        i+=1

    return clusters


# Esta segunda funcion se encarga de contar los elementos de cada cluster, para asi obtener el
# mas repetedio o predicho en cada cluster
def get_clusters_count_labels(clusters):
    # Se itera los clusters y se obtiene en forma de diccionario
    predictions = []
    for cluster in clusters:
        predictions += [dict(Counter(cluster))]

    return predictions

# Esta otra funcion, lo que hace es guardar la clase actual con la clase predicha por cada elemento
# de cada cluster
def set_actual_and_predicted_pair(clusters):
    classes = [[] for _ in range(len(clusters))]

    cluster_idx = 0
    for cluster in clusters:
        # Se obtiene la prediccion de ese cluster
        pred_class = Counter(cluster).most_common(1)[0][0]
        # Se iteran entre los elementos para guardar el par (actual, predicho)
        for actual_class in cluster:
            classes[cluster_idx].append((pred_class, actual_class))
        cluster_idx += 1

    return classes

# Finalmente, la funcion que va a imprimir la matriz de confusion pasando las dos listas
# con las clases predichas y las actuales, para ver las diferencias

def get_confusion_matrix(clusters):
    predicts = []
    actuals = []

    # Se guardan en esas dos listas todas las predicciones y los actuales
    for cluster in clusters:
        for pair in cluster:
            predicts.append(pair[0])
            actuals.append(pair[1])

    # Se dibuja la matriz de confusion
    confusion_matrix = metrics.confusion_matrix(actuals, predicts)

    cm_display = metrics.ConfusionMatrixDisplay(
        confusion_matrix=confusion_matrix, display_labels=[1, 2, 3])

    cm_display.plot()
    plt.show()

```

# K = 3

```python
clusters = get_predicted_cluster_labels(km_3_classifier.predict(X_iris), 3)
print(get_clusters_count_labels(clusters))

get_confusion_matrix(set_actual_and_predicted_pair(clusters))
```

Como se puede apreciar, de nuevo, hay una flor la cual esta muy bien clasificada, al 100% de acierto dentro de un cluster, mientras que las otras 2, tienen un error, siendo igual que pasaba con la implementacion propia.
